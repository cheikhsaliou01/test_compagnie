# Creer une colonne tranche d'age < 18 entre 18 et 25 , de 25 a 50 et pkus de 50 

```python
import numpy as np 
data["tranche_age"] = np.nan
data["tranche_age"].head()
```

```python
data["tranche_age"] [data["age"] < 18] = "<18"
data["tranche_age"] [(data["age"] >= 18) & (data["age"] < 25)] = "<18-25"
data["tranche_age"] [(data["age"] >= 25) & (data["age"] < 50)] = "<25-50"
data["tranche_age"] [(data["age"] >= 50)] = ">=50"
```
